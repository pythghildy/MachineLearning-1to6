```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
print(boston['feature_names'])
```

```python
print(boston['DESCR'])
```

```python
bostondf = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
bostondf.head()
```

```python
boston.target.shape
```

```python
bostondf['Price'] = boston.target
```

```python
bostondf.head()
```

```python
bostondf.describe()
```

```python
bostondf.isnull().sum()
```

```python
sns.distplot(bostondf['Price'])
```

There seems outliers. We'll do box plot

```python
sns.boxplot(x='Price',data=bostondf)
```

Price 38 and above are Outliers

```python
(bostondf[bostondf['Price']>38].shape[0]/bostondf.shape[0])*100
```

6.5% outliers. Can be dropped

```python
bostondf.shape
```

```python
Price_Outliers = bostondf[bostondf['Price']>38]
bostondf = bostondf.drop(Price_Outliers.index,axis=0)
bostondf.shape
```

```python
sns.pairplot(bostondf)
```

```python
X = bostondf.drop('Price', axis=1)
y = bostondf['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
rfc = RandomForestRegressor(n_jobs=-1)
```

```python
rfc.fit(X_train, y_train)
```

```python
prediction = rfc.predict(X_test)
```

```python
print('The Mean Sqaured Error is : ', mean_squared_error(prediction,y_test))
print('The Mean Absolute Error is : ', mean_absolute_error(prediction,y_test))
print('The Model accuracy/r2 Score using Random Decision Tree Algorithm is ', r2_score(prediction,y_test))
```

# Model Accuracy is 81.71%
